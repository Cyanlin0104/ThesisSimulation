```python
import numpy as np
from matplotlib import pyplot as plt
```

$$  w = \frac{1}{x^4} $$ 

```python
w = lambda x : 1.0 / (x**4)
d = np.linspace(1e-5, 1, 100)

fig, axes = plt.subplots()
axes.plot(d, w(d))
axes.set_ylim([0, 1000])
```

$$ u(x) = 1 - e^{\frac{-x^2}{2\sigma^2}} $$

```python
sigma = 0.01
def u(x_dot): 
    return 1 - np.exp(-x_dot**2/(2*sigma**2))

d_dot = np.linspace(-1, 1, 1000)

fig, axes = plt.subplots()

axes.plot(d_dot, u(d_dot))
axes.axvline(x=0, linestyle='--')
axes.set_ylim([-1, 2])
axes.set_xlabel(r'$\dot{s}$')
```

$$ s(q) = \frac{q - l_l}{l_u - l_l}$$
$$ d(q) = 4s(q)(1 - s(q))$$

```python
ll = -3
lu = 3

def d(x):
    s = (x - ll) / (lu - ll)
    return 4*s*(1-s)

fig, axes = plt.subplots()
x = np.linspace(-4, 4, 800)
axes.plot(x, d(x))
```

$$ h^{\alpha}(x) = \frac{1}{\alpha}\log{(e^{\alpha x} + e^{-\alpha x})}$$

```python
alpha = 1
def h(x):
    return np.log(np.exp(alpha*x) + np.exp(-alpha*x))/alpha

# equivalent to h
def h1(x):
    return x + (np.log(1 + np.exp(-2*alpha*x)) / alpha)

fig, axes = plt.subplots()
x = np.linspace(0, 100, 1000)
axes.plot(x, h1(x))
```
